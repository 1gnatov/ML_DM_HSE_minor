#### Описание признаков в таблице

- `match_id`: идентификатор матча в наборе данных
- `start_time`: время начала матча (unixtime)
- `lobby_type`: тип комнаты, в которой собираются игроки (расшифровка в `dictionaries/lobbies.csv`)
- Наборы признаков для каждого игрока (игроки команды Radiant — префикс `rN`, Dire — `dN`):
    - `r1_hero`: герой игрока (расшифровка в dictionaries/heroes.csv)
    - `r1_level`: максимальный достигнутый уровень героя (за первые 5 игровых минут)
    - `r1_xp`: максимальный полученный опыт
    - `r1_gold`: достигнутая ценность героя
    - `r1_lh`: число убитых юнитов
    - `r1_kills`: число убитых игроков
    - `r1_deaths`: число смертей героя
    - `r1_items`: число купленных предметов
- Признаки события "первая кровь" (first blood). Если событие "первая кровь" не успело произойти за первые 5 минут, то признаки принимают пропущенное значение
    - `first_blood_time`: игровое время первой крови
    - `first_blood_team`: команда, совершившая первую кровь (0 — Radiant, 1 — Dire)
    - `first_blood_player1`: игрок, причастный к событию
    - `first_blood_player2`: второй игрок, причастный к событию
- Признаки для каждой команды (префиксы `radiant_` и `dire_`)
    - `radiant_bottle_time`: время первого приобретения командой предмета "bottle"
    - `radiant_courier_time`: время приобретения предмета "courier" 
    - `radiant_flying_courier_time`: время приобретения предмета "flying_courier" 
    - `radiant_tpscroll_count`: число предметов "tpscroll" за первые 5 минут
    - `radiant_boots_count`: число предметов "boots"
    - `radiant_ward_observer_count`: число предметов "ward_observer"
    - `radiant_ward_sentry_count`: число предметов "ward_sentry"
    - `radiant_first_ward_time`: время установки командой первого "наблюдателя", т.е. предмета, который позволяет видеть часть игрового поля
- Итог матча (данные поля отсутствуют в тестовой выборке, поскольку содержат информацию, выходящую за пределы первых 5 минут матча)
    - `duration`: длительность
    - `radiant_win`: 1, если победила команда Radiant, 0 — иначе
    - Состояние башен и барраков к концу матча (см. описание полей набора данных)
        - `tower_status_radiant`
        - `tower_status_dire`
        - `barracks_status_radiant`
        - `barracks_status_dire`

**Замените пропуски в признаке first_blood_team на -1, бинаризуйте и выкиньте исходный признак.**

```python
for data in [dota_train, dota_test]:
    data['first_blood_team'].fillna(-1, inplace=True)

    data['first_blood_radiant'] = data['first_blood_team'].apply(
        lambda x: 1 if x == 1 else 0)
    data['first_blood_unknown'] = data['first_blood_team'].apply(
        lambda x: 1 if x == -1 else 0)

    data.drop('first_blood_team', axis=1, inplace=True)
```

**Аггрегируйте признаки 'level', 'xp', 'gold', 'lh', 'kills', 'deaths' и 'items' в разности этих признаков для двух команд.**

```python
for data in [dota_train, dota_test]:
    for aggr_feat_postfix in ['level', 'xp', 'gold', 'lh', 'kills',
                                 'deaths', 'items']:
        for team_prefix in ['r', 'd']:
            data['aggr_' + team_prefix + '_' + aggr_feat_postfix] = \
                    data[team_prefix + '1_' + aggr_feat_postfix]
            for i in range(2,6):
                data['aggr_' + team_prefix + '_' + aggr_feat_postfix] += \
                    data[team_prefix + str(i) + '_' + aggr_feat_postfix]
        data['aggr_' + aggr_feat_postfix + '_diff'] = \
                data['aggr_r_' + aggr_feat_postfix] -\
                data['aggr_d_' + aggr_feat_postfix]
```

**Реализуйте функцию my_cross_validation. Она должна принимать словарь со значениями параметров и для всех значений параметра  cv_param_name (который хотим подобрать) проводить KFold кросс-валидацию, записывая ROC AUC.**

```python
from sklearn.cross_validation import KFold
from sklearn.metrics import roc_auc_score
```

```python
def my_cross_validation(train_data, train_labels, n_folds, classifier, param_dic, 
                        cv_param_name, fold_random_state=1,
                        verbose=True):
    training_times = []
    mean_cv_scores = []
    cv_param_values = param_dic[cv_param_name]

    for param_value in param_dic[cv_param_name]:
        scores = []
        init_time_before_cv = time()
        folds = KFold(n=train_labels.shape[0], n_folds=n_folds, 
                      shuffle=True, random_state=fold_random_state)
        for train_idx, test_idx in folds:
            X_train, X_test = train_data.iloc[train_idx], train_data.iloc[test_idx]
            y_train, y_test = train_labels.iloc[train_idx], train_labels.iloc[test_idx]
            
            clf_params = param_dic
            clf_params[cv_param_name] = param_value
            classifier.set_params(**clf_params)
            classifier.fit(X_train, y_train)
            scores.append(roc_auc_score(y_test, classifier.predict_proba(X_test)[:,1]))

        time_passed = time() - init_time_before_cv
        training_times.append(time_passed)    
        mean_cv_scores.append(np.mean(scores))
        if verbose:
            print("{0}: {1}, average CV ROC AUC: {2}, time: {3}".format(cv_param_name,
                                                                        param_value, 
                                                                        round(np.mean(scores), 4),
                                                                        round(time_passed, 4)))
    return cv_param_values, mean_cv_scores, training_times
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
def train_test_time_series(train_and_test, n_folds, classifier, param_dic,
                           cv_param_name, window=True, window_size=10000,
                          progress_widget=False, scale=True):
    cv_scores = []
    train_labels = train_and_test[train_and_test['radiant_win'] != -1]['radiant_win']
    train_ids = train_and_test[train_and_test['radiant_win'] != -1].index
    test_ids = train_and_test[train_and_test['radiant_win'] == -1].index
    
    valid_pred = pd.DataFrame([0] * len(train_ids))
    valid_pred.index = train_ids
    predictions = pd.DataFrame([0] * len(test_ids))
    predictions.index = test_ids
    
    folds = KFold(n=train_and_test.shape[0], n_folds=n_folds)
    for _, test_idx in folds:
        if not np.min(test_idx):
            current_fold = train_and_test.iloc[test_idx]
            current_fold_train = current_fold[current_fold['radiant_win'] != -1]
            current_fold_test = current_fold[current_fold['radiant_win'] == -1]
            X_train = current_fold_train.drop('radiant_win', axis=1)
            X_valid = X_train.copy()
            y_train = current_fold_train['radiant_win']
            y_valid = y_train.copy()
            X_test = current_fold_test.drop('radiant_win', axis=1) 
        else:
            if window:
                history_idx = range(np.min(test_idx) - window_size, np.min(test_idx))
            else:
                history_idx = range(np.min(test_idx))
            history = train_and_test.iloc[history_idx]
            current_fold = train_and_test.iloc[test_idx]
            history_train = history[history['radiant_win'] != -1]
            y_train = history_train['radiant_win']

            current_valid = current_fold[current_fold['radiant_win'] != -1]
            current_test = current_fold[current_fold['radiant_win'] == -1]

            y_valid = current_valid['radiant_win']
            X_train = history_train.drop('radiant_win', axis=1)
            X_valid = current_valid.drop('radiant_win', axis=1)
            X_test = current_test.drop('radiant_win', axis=1)

        print(X_train.index.min(), X_train.index.max())
        print(X_valid.index.min(), X_valid.index.max())
        print(X_test.index.min(), X_test.index.max())
        
        
        if scale:
            scaler = StandardScaler() 
            X_train_scaled = scaler.fit_transform(X_train)
            X_valid_scaled = scaler.fit_transform(X_valid)
            X_test_scaled = scaler.fit_transform(X_test)
        else:
            X_train_scaled = X_train.copy()
            X_valid_scaled = X_valid.copy()
            X_test_scaled = X_test.copy()
        
        best_roc_auc, best_param_value = 0, 0
        
        for param_value in param_dic[cv_param_name]:

            clf_params = param_dic.copy()
            clf_params[cv_param_name] = param_value
            classifier.set_params(**clf_params)
            classifier.fit(X_train_scaled, y_train)

            current_roc = roc_auc_score(y_valid, 
                            classifier.predict_proba(X_valid_scaled)[:,1])

            if current_roc > best_roc_auc:
                best_roc_auc = current_roc
                best_param_value = param_value

        clf_params[cv_param_name] = best_param_value
        classifier.set_params(**clf_params)
        classifier.fit(X_train_scaled, y_train)
        
        print('Best param', best_param_value)

        valid_pred.loc[X_valid.index] = \
                (classifier.predict_proba(X_valid_scaled)[:,1])\
                .reshape([X_valid.shape[0], 1])
        predictions.loc[X_test.index] = \
            (classifier.predict_proba(X_test_scaled)[:,1])\
            .reshape([X_test.shape[0], 1])
        
        print('Val score', roc_auc_score(y_valid, valid_pred.loc[X_valid.index]))
        if np.min(test_idx):
            cv_scores.append(roc_auc_score(y_valid, valid_pred.loc[X_valid.index]))            
    return predictions, np.mean(cv_scores), valid_pred
```

```python
rf_pred, rf_cv_score, rf_valid_pred = train_test_time_series(sub_train_and_test, 10, 
                                                  RandomForestClassifier(),
                      param_dic={'n_estimators': 150, 'max_depth': range(5, 15, 2), 
                                 'random_state': 42, 'n_jobs': -1},
                      cv_param_name='max_depth', window=False, window_size=10000, scale=False)
```
