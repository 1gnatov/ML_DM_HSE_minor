# Семинар 11. Линейный и бинарный поиск 

## Часть 1. Линейный поиск

Когда элементы данных хранятся коллекцией в виде **списка**, мы говорим, что между ними **линейные или последовательные** отношения. Каждый элемент хранится на определённой позиции относительно прочих. В списках Python она задаётся индексом данного элемента. Поскольку значения индексов упорядочены, мы имеем возможность последовательно проходить по ним. Этот процесс приводит к нашей первой поисковой технике - **последовательному поиску**:

<img src = '../img/seqsearch.png'>

Начиная с первого элемента в списке, мы просто движемся от значения к значению, следуя внутреннему порядку последовательности, до тех пор, пока:

1. либо не найдём то, что ищем 
2. либо не достигнем последнего элемента. 
3. последовательность не содержит искомое значение



```python
def sequentialSearch(alist, item):
    pos = 0
    found = False

    while pos < len(alist) and not found:
        if alist[pos] == item:
            found = True
        else:
            pos = pos+1

    return found

testlist = [1, 2, 32, 8, 17, 19, 42, 13, 0]
print(sequentialSearch(testlist, 3))
print(sequentialSearch(testlist, 13))
```

### Часть 2. Бинарный поиск

Существует возможность использовать преимущества упорядоченного списка, если правильно сравнения. В последовательном поиске, когда мы сравниваем первый элемент, может быть до n−1 элемента, которые нужно ещё просмотреть, если первый - не то, что мы ищем.

**Бинарный поиск** начинает проверять элементы с находящегося в середине. Если это то, что нам нужно, - всё готово. Если нет, то можно использовать упорядоченную природу списка, исключив половину оставшихся элементов.

<img src = '../img/binsearch.png'>

```python
def binarySearch(alist, item):
    first = 0
    last = len(alist)-1
    found = False

    while first<=last and not found:
        midpoint = (first + last)//2
        if alist[midpoint] == item:
            found = True
        else:
            if item < alist[midpoint]:
                last = midpoint-1
            else:
                first = midpoint+1

    return found

testlist = [0, 1, 2, 8, 13, 17, 19, 32, 42,]
print(binarySearch(testlist, 3))
print(binarySearch(testlist, 13))
```
